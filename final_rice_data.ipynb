```python
df.drop('production', axis = 1, inplace = True)
df.head()
```

```python
df.drop('harvested', axis = 1, inplace = True)
df.head()
```

```python
col = ['year', 'state', 'precipitation', 'temperature', 'planted', 'yield']
```

```python
df = df[col]
```

```python
df.head()
```

```python
#Preparing the dataframe for machine learning model
# Dropping 'Yield' column and creating new dataframe for input of the model
#axis =1 means to drop column not row
x = df.drop('yield', axis = 1)

#create series with yield. This is our target for supervised learning
y = df['yield']
```

```python
x.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0, shuffle = True)
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler
```

```python
ohe = OneHotEncoder(drop = 'first')
scale = StandardScaler()

preprocessor = ColumnTransformer(
    transformers = [
        ('StandardScale', scale, [2, 3, 4]),
        ('OneHotEncode', ohe, [1])
    ],
    remainder = 'passthrough'
)
    
    
```

```python
x_train_dummy = preprocessor.fit_transform(x_train)
x_test_dummy = preprocessor.fit_transform(x_test)
```

```python
x_train_dummy
```

```python
x_test_dummy
```

```python
preprocessor.get_feature_names_out(col[:-1])
```

```python
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_absolute_error, r2_score, accuracy_score
```

```python
models = {
    'Linear Regression': LinearRegression(),
    'Lasso': Lasso(),
    'Ridge': Ridge(),
    'Decision Tree': DecisionTreeRegressor(),
    'KNN': KNeighborsRegressor(),
}

for name, md in models.items():
    md.fit(x_train_dummy, y_train)
    y_pred = md.predict(x_test_dummy)
    print(f"{name}: mae : {mean_absolute_error(y_test, y_pred)} score : {r2_score(y_test, y_pred)}")
```

```python
dtr = DecisionTreeRegressor()
dtr.fit(x_train_dummy, y_train)
dtr.predict(x_test_dummy)
```

```python
def prediction(year, state, precipitation, temperature, planted):
    features = np.array([[year, state, precipitation, temperature, planted]])
    transform_features = preprocessor.transform(features)
    predicted_yield = dtr.predict(transform_features).reshape(-1, 1)
    return predicted_yield[0][0]
```
